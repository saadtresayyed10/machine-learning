# Import Library

```python
import pandas as pd
```

# Load Dataset

```python
df = pd.read_csv("dirty_cafe_sales.csv")
df.head(20)
```

```python
df.info()
```

# Getting all object dtypes columns

```python
dfCols = df.select_dtypes(include="object").columns
dfCols
```

```python
for i in dfCols:
    print(i)
```

```python
# Imputing data using mode
```

```python
for i in dfCols:
    df.fillna(df[i].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```
